```python
#Enhancement to for loops
# automating the print statements:
```

```python
my_students=['srinivasa','sandhya','nidhi','prasad','suhas','phani']
```

```python
print(my_students)
```

```python
#automate the print statement for every element in the list:
```

```python
for x in my_students:
    print(f"{x}, good job guys!!! Taks completed on time")
```

```python
for x in my_students:
    print(f"{x.title()}, good job guys!!! Taks completed on time")
```

```python
for x in my_students:
    print(f"{x.upper()}, good job guys!!! Taks completed on time")
```

```python
for x in my_students:
    print(f"\n {x.upper()}, good job guys!!! Taks completed on time")
```

```python
# further enhancements
```

```python
for x in my_students:
    print(f"Thanks for completing the job on time, {x.title()}")
print("\n Thank you all")    
```
